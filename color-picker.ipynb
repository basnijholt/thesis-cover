```python
import matplotlib.pyplot as plt
import numpy as np

options = [
    "data/gradient-sweep-alpha2/data_learner_0587.pickle",
    "data/gradient-sweep-angle-0-45/data_learner_0334.pickle",
    "data/gradient-sweep-alpha/data_learner_0027.pickle",
    "data/gradient-sweep-alpha/data_learner_0190.pickle",
    "data/gradient-sweep-alpha/data_learner_0184.pickle",
    "data/mu-sweep/data_learner_0051.pickle",
    "data/gradient-sweep-angle-0-45/data_learner_0057.pickle",
    "data/gradient-sweep-alpha2/data_learner_0584.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0233.pickle",
    "data/gradient-sweep-alpha/data_learner_0030.pickle",
    "data/gradient-sweep-alpha/data_learner_0187.pickle",
    "data/gradient-sweep-angle-0-45/data_learner_0069.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0811.pickle",
    "data/gradient-sweep-rotation-0-45/data_learner_0016.pickle",
    "data/mu-sweep2/data_learner_0199.pickle",
    "data/gradient-sweep-angle-0-45/data_learner_0068.pickle",
    "data/gradient-sweep-angle-0-45/data_learner_0332.pickle",
    "data/mu-sweep2/data_learner_0200.pickle",
    "data/gradient-sweep-alpha2/data_learner_0379.pickle",
    "data/mu-sweep2/data_learner_0201.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0035.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0021.pickle",
    "data/gradient-sweep-angle-0-45/data_learner_0325.pickle",
    "data/gradient-sweep-alpha/data_learner_0220.pickle",
    "data/gradient-sweep-alpha2/data_learner_0178.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0723.pickle",
    "data/gradient-sweep-alpha2/data_learner_0080.pickle",
    "data/mu-sweep2/data_learner_0259.pickle",
    "data/gradient-sweep-alpha/data_learner_0118.pickle",
    "data/mu-sweep2/data_learner_0113.pickle",
    "data/gradient-sweep-alpha/data_learner_0119.pickle",
    "data/mu-sweep2/data_learner_0258.pickle",
    "data/gradient-sweep-alpha2/data_learner_0444.pickle",
    "data/gradient-sweep-alpha/data_learner_0086.pickle",
    "data/gradient-sweep-alpha/data_learner_0284.pickle",
    "data/gradient-sweep-alpha/data_learner_0286.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0052.pickle",
    "data/gradient-sweep-alpha2/data_learner_0083.pickle",
    "data/gradient-sweep-alpha/data_learner_0087.pickle",
    "data/mu-sweep2/data_learner_0111.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0457.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0709.pickle",
    "data/gradient-sweep-angle-0-45/data_learner_0345.pickle",
    "data/mu-sweep2/data_learner_0049.pickle",
    "data/gradient-sweep-alpha/data_learner_0254.pickle",
    "data/gradient-sweep-alpha2/data_learner_0092.pickle",
    "data/mu-sweep/data_learner_0183.pickle",
    "data/gradient-sweep-alpha2/data_learner_0079.pickle",
    "data/mu-sweep2/data_learner_0115.pickle",
    "data/mu-sweep2/data_learner_0101.pickle",
    "data/mu-sweep2/data_learner_0114.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0693.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0485.pickle",
    "data/gradient-sweep-alpha/data_learner_0094.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0726.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0685.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0691.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0684.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0492.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0479.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0727.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0055.pickle",
    "data/gradient-sweep-alpha/data_learner_0059.pickle",
    "data/gradient-sweep-alpha2/data_learner_0076.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0879.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0461.pickle",
    "data/mu-sweep2/data_learner_0245.pickle",
    "data/mu-sweep2/data_learner_0047.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0703.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0715.pickle",
    "data/gradient-sweep-rotation-0-45/data_learner_0082.pickle",
    "data/gradient-sweep-rotation-0-45/data_learner_0083.pickle",
    "data/gradient-sweep-alpha2/data_learner_0921.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0710.pickle",
    "data/gradient-sweep-alpha2/data_learner_0272.pickle",
    "data/gradient-sweep-alpha/data_learner_0088.pickle",
    "data/mu-sweep2/data_learner_0256.pickle",
    "data/gradient-sweep-alpha/data_learner_0116.pickle",
    "data/mu-sweep2/data_learner_0257.pickle",
    "data/gradient-sweep-alpha/data_learner_0062.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0711.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0705.pickle",
    "data/gradient-sweep-rotation-0-45-move-SO-and-potential/data_learner_0013.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0713.pickle",
    "data/gradient-sweep-alpha/data_learner_0276.pickle",
    "data/gradient-sweep-alpha2/data_learner_0073.pickle",
    "data/gradient-sweep-alpha/data_learner_0060.pickle",
    "data/mu-sweep2/data_learner_0255.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0459.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0471.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0699.pickle",
    "data/gradient-sweep-alpha/data_learner_0061.pickle",
    "data/gradient-sweep-angle-0-45/data_learner_0372.pickle",
    "data/mu-sweep2/data_learner_0233.pickle",
    "data/gradient-sweep-alpha2/data_learner_0175.pickle",
    "data/mu-sweep2/data_learner_0192.pickle",
    "data/mu-sweep2/data_learner_0179.pickle",
    "data/gradient-sweep-angle-0-45/data_learner_0076.pickle",
    "data/mu-sweep2/data_learner_0232.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0748.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0006.pickle",
    "data/gradient-sweep-alpha2/data_learner_0566.pickle",
    "data/gradient-sweep-alpha2/data_learner_0823.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0830.pickle",
    "data/gradient-sweep-angle-0-45/data_learner_0049.pickle",
    "data/gradient-sweep-alpha2/data_learner_0188.pickle",
    "data/mu-sweep2/data_learner_0037.pickle",
    "data/gradient-sweep-alpha/data_learner_0028.pickle",
    "data/mu-sweep2/data_learner_0235.pickle",
    "data/mu-sweep2/data_learner_0180.pickle",
    "data/gradient-sweep-alpha2/data_learner_0370.pickle",
    "data/mu-sweep2/data_learner_0181.pickle",
    "data/gradient-sweep-rotation-0-45-move-SO-and-potential/data_learner_0489.pickle",
    "data/gradient-sweep-alpha/data_learner_0029.pickle",
    "data/mu-sweep2/data_learner_0036.pickle",
    "data/mu-sweep2/data_learner_0034.pickle",
    "data/gradient-sweep-alpha/data_learner_0215.pickle",
    "data/gradient-sweep-rotation-0-90-move-SO-slowly/data_learner_0228.pickle",
    "data/mu-sweep2/data_learner_0236.pickle",
    "data/mu-sweep2/data_learner_0222.pickle",
    "data/gradient-sweep-alpha/data_learner_0188.pickle",
    "data/mu-sweep2/data_learner_0035.pickle",
]

cmaps = [
    (
        "Perceptually Uniform Sequential",
        ["viridis", "plasma", "inferno", "magma", "cividis"],
    ),
    (
        "Sequential",
        [
            "Greys",
            "Purples",
            "Blues",
            "Greens",
            "Oranges",
            "Reds",
            "YlOrBr",
            "YlOrRd",
            "OrRd",
            "PuRd",
            "RdPu",
            "BuPu",
            "GnBu",
            "PuBu",
            "YlGnBu",
            "PuBuGn",
            "BuGn",
            "YlGn",
        ],
    ),
    (
        "Sequential (2)",
        [
            "binary",
            "gist_yarg",
            "gist_gray",
            "gray",
            "bone",
            "pink",
            "spring",
            "summer",
            "autumn",
            "winter",
            "cool",
            "Wistia",
            "hot",
            "afmhot",
            "gist_heat",
            "copper",
        ],
    ),
    (
        "Diverging",
        [
            "PiYG",
            "PRGn",
            "BrBG",
            "PuOr",
            "RdGy",
            "RdBu",
            "RdYlBu",
            "RdYlGn",
            "Spectral",
            "coolwarm",
            "bwr",
            "seismic",
        ],
    ),
    ("Cyclic", ["twilight", "twilight_shifted", "hsv"]),
    (
        "Qualitative",
        [
            "Pastel1",
            "Pastel2",
            "Paired",
            "Accent",
            "Dark2",
            "Set1",
            "Set2",
            "Set3",
            "tab10",
            "tab20",
            "tab20b",
            "tab20c",
        ],
    ),
    (
        "Miscellaneous",
        [
            "flag",
            "prism",
            "ocean",
            "gist_earth",
            "terrain",
            "gist_stern",
            "gnuplot",
            "gnuplot2",
            "CMRmap",
            "cubehelix",
            "brg",
            "gist_rainbow",
            "rainbow",
            "jet",
            "nipy_spectral",
            "gist_ncar",
        ],
    ),
]

all_cmaps = [f"{k}: {cm}" for k, _cmaps in cmaps for cm in _cmaps]



def plot_color_gradients(cmap_category, cmap_list):
    # Create figure and adjust figure height to number of colormaps

    nrows = len(cmap_list)
    figh = 0.35 + 0.15 + (nrows + (nrows - 1) * 0.1) * 0.22
    fig, axes = plt.subplots(nrows=nrows, figsize=(6.4, figh))
    fig.subplots_adjust(top=1 - 0.35 / figh, bottom=0.15 / figh, left=0.2, right=0.99)

    axes[0].set_title(cmap_category + " colormaps", fontsize=14)

    gradient = np.linspace(0, 1, 256)
    gradient = np.vstack((gradient, gradient))
    for ax, name in zip(axes, cmap_list):
        ax.imshow(gradient, aspect="auto", cmap=plt.get_cmap(name))
        ax.text(
            -0.01,
            0.5,
            name,
            va="center",
            ha="right",
            fontsize=10,
            transform=ax.transAxes,
        )

    # Turn off *all* ticks & spines, not just the ones with colormaps.
    for ax in axes:
        ax.set_axis_off()

for cmap_category, cmap_list in cmaps:
    plot_color_gradients(cmap_category, cmap_list)

plt.show()
```

```python
import adaptive
import matplotlib.colors as colors
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import FloatSlider, interact, IntSlider
from thesis_cover import load_learner

x_size, y_size = xy_size = (14.13, 9.84)


def plot(fname, cmap, min_clip=0.0, max_clip=1.0, exp=1.0, invert=False):
    learner = load_learner(fname)
    xs, ys, zs = data = learner.interpolated_on_grid(1000)
    cmap = cmap.split(": ")[-1] + ("_r" if invert else "")
    fcmap = getattr(plt.cm, cmap)
    mymap = colors.LinearSegmentedColormap.from_list(
        "my_colormap", fcmap(np.linspace(min_clip, max_clip, 256) ** exp),
    )
    zs = mymap(zs)
    fig, ax = plt.subplots(figsize=xy_size)
    im = ax.imshow(
        np.rot90(zs),
        extent=(-0.5 * x_size, 0.5 * x_size, -0.5 * y_size, 0.5 * y_size),
        zorder=3,
    )
    plt.show()


interact(
    plot,
    fname=options,
    cmap=all_cmaps,
    min_clip=FloatSlider(min=0, max=1),
    max_clip=FloatSlider(value=1, min=0, max=1),
    exp=FloatSlider(value=1, min=0.1, max=3),
)
```
